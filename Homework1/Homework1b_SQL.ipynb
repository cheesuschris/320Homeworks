Problem 6:

Create a dataframe with averages of base pay, benefits, and overtime for each job title, as well as a column with the average of these three values. Use `JobTitle`, `AvgBasePay`, `AvgBenefits`, `AvgOvertimePay`, `AvgTotal` as the column name respectively. Use `df6` as the pandas DataFrame variable to store your query result

```python
query = 'SELECT JobTitle, AVG(COALESCE(BasePay, 0)) AS AvgBasePay, AVG(COALESCE(Benefits, 0)) AS AvgBenefits, AVG(COALESCE(OvertimePay, 0)) AS AvgOvertimePay, (AVG(COALESCE(BasePay, 0)) + AVG(COALESCE(Benefits, 0)) + AVG(COALESCE(OvertimePay, 0))) / 3.0 AS AvgTotal FROM Salaries GROUP BY JobTitle'
df6 = pd.read_sql(query, conn)
df6.head()
```

# Table Creation

Problem 7:

Now we'll create our own table in our database. Separate the Salaries table by Year, and add it back to the database. (You may use basic python to complete the task, however, should still use SQL to query the data.)

Use `Y` + `Year` as the table name (i.e. `Y2011`).

Use `df7_list` as a list variable of dataframe to save the data. For example, `df7_list[0]` will represent Year 2011's data.


```python
df7_list = []
for y in ['2011','2012','2013','2014']:
    query = f'SELECT * FROM Salaries WHERE Year={y}'
    df7 = pd.read_sql(query, conn)
    df7.to_sql(name='Y'+y, con=conn, if_exists='replace')
    df7_list.append(df7)

df7_list[0].head()
```

Problem 11:

Which directors have the highest overall voting average? - show the top 5 directors' name and their average rating. Your result should have a column `ave_rating` and a column `director_name`. Use `df11` as the pandas DataFrame variable to store your query result

```python
query = 'SELECT AVG(movies.vote_average) AS avg_rating, directors.name AS director_name FROM movies LEFT JOIN directors ON directors.id = movies.director_id GROUP BY directors.name ORDER BY avg_rating DESC LIMIT 5'
df11 = pd.read_sql(query, conn)
df11.head()
```

Problem 12:

What are the top five directors by how much their average budget is? Your result should have a column `ave_budget` and a column `director_name`. Use `df12` as the pandas DataFrame variable to store your query result

```python
# Problem 12: Top 5 directors by average budget
query = 'SELECT AVG(movies.budget) AS avg_budget, directors.name AS director_name FROM movies LEFT JOIN directors ON directors.id = movies.director_id GROUP BY directors.name ORDER BY avg_budget DESC LIMIT 5'
df12 = pd.read_sql(query, conn)
df12.head()
```
